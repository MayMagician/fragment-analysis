```python
from google.colab import drive
import pandas as pd
import igraph as ig
import numpy as np
import random

drive.mount('/content/drive')

df = pd.read_csv('/content/drive/MyDrive/Fragment/fragmentTry/df_step5.csv')
edges_df = pd.read_csv('/content/drive/MyDrive/Fragment/fragmentTry/graph_step5.edgelist', sep=' ', names=['source', 'target', 'weight'])
g = ig.Graph.TupleList(edges_df.itertuples(index=False), weights=True)
node_names = g.vs['name']

print(f"{g.vcount()} nodes, {g.ecount()} edges")
```

# Leiden

```python
random.seed(42)
np.random.seed(42)

partition = g.community_leiden(weights='weight', objective_function='modularity')

print(f"number of social communities: {len(partition)}")
print(f"modularity score: {g.modularity(partition.membership, weights='weight'):.4f}")

#add results to df
results = {node['name']: partition.membership[i] for i, node in enumerate(g.vs)} #key = asset name, value = community id
df['comm_louvain'] = df['Asset'].map(results)
```
